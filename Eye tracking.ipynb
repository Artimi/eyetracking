## Code
Here you can see the code that was used to analyze data.

```python
from __future__ import division
import csv
import numpy as np
import sklearn.cluster
import itertools
from collections import namedtuple

FILE_NAME = 's1-s6-final.csv'
DELTA_T = 0.001  #sampling rate 1000 Hz
LEN_FIXATIONS_THRESHOLD = 10 * DELTA_T  #threshold for short fixations that are probably incorectly sorted 

SubjectResult = namedtuple("SubjectResult", ['subject', "MFD_true", "MFD_SD_true", "MFD_false", "MFD_SD_false",
                           "MSA_true", "MSA_SD_true", "MSA_false", "MSA_SD_false"])
OverallResult = namedtuple('OverallResult', ['MFD_overall_true', 'MFD_overall_true_SD',
                                             'MFD_overall_false', 'MFD_overall_false_SD',
                                             'MSA_overall_true', 'MSA_overall_true_SD',
                                             'MSA_overall_false', 'MSA_overall_false_SD'])

class Record(object):
    '''
    Object that stores all position data and computes all statistics like velocity, peaks, 
    length of fixations, Mean Fixation Duration (MFD) and Mean Saccade Amplitude.
    '''
    def __init__(self, line):
        self.subject = line[0]
        self.known = line[1]
        self.position = np.array(map(float, line[2:]))
        self.position = np.reshape(self.position, (2, len(self.position) // 2), order='F')
        self.velocity = self.compute_velocity(self.position)
        self.peaks = self.get_peaks(self.velocity)
        self.len_fixations = self.get_len_fixations(self.peaks)
        self.MFD = self.get_MFD()
        self.MSA = self.get_MSA()

    def compute_velocity(self, position):
        '''
        From the coordinates compute velocity array
        '''
        velocity = np.zeros([position.shape[1] - 1], dtype=float)
        for i in range(1, position.shape[1]):
            velocity[i - 1] = np.sqrt((position[0, i] - position[0, i - 1]) ** 2
                                  + (position[1, i] - position[1, i - 1]) ** 2) / DELTA_T
        return velocity

    def get_peaks(self, velocity):
        '''
        Using KMeans algorithm split velocity to two clusters: fixations and saccades (peaks).
        Assure peaks (saccades) have label True.
        '''
        k_means = sklearn.cluster.KMeans(n_clusters=2)
        k_means.fit(np.expand_dims(velocity, 1))
        labels = k_means.labels_
        clusters = k_means.cluster_centers_
        labels = labels.astype(bool)
        if clusters[0, 0] > clusters[1, 0]:
            labels = np.logical_not(labels)
        return labels

    def get_len_fixations(self, peaks):
        '''
        Return array of lengths of fixations in seconds.
        '''
        len_fixations = np.array([])
        for bit, group in itertools.groupby(peaks):
            if not bit:
                len_fixations = np.append(len_fixations, sum(1 for _ in group) * DELTA_T)
        return len_fixations

    def get_MSA(self):
        '''
        Computes Mean Saccade Amplitude. For every saccade pick the maximum value(amplitude)
        and average them all.
        '''
        SAs = np.array([])
        offset = 0
        for bit, group in itertools.groupby(self.peaks):
            group_length = sum(1 for _ in group)
            if bit:
                SAs = np.append(SAs, np.max(self.velocity[offset:offset + group_length]))
            offset += group_length
        return np.average(SAs)

    def get_MFD(self):
        '''
        Computes Mean Fixation Duration. Average length of all fixations above threshold.
        '''
        return np.average(self.len_fixations[self.len_fixations > LEN_FIXATIONS_THRESHOLD])


def get_record(file_path=FILE_NAME):
    '''
    Iterator that returns processed Records
    '''
    with open(file_path) as csvfile:
        for line in csv.reader(csvfile):
            yield Record(line)


def get_result_dict():
    '''
    Returns dictionary that contains all MSA and MFD according to subject nad known state.
    '''
    result = {}
    for rec in get_record():
        try:
            result[rec.subject]
        except KeyError:
            result[rec.subject] = {"true":
                                   {"MSA": np.array([]),
                                    "MFD": np.array([])},
                                   "false":
                                   {"MSA": np.array([]),
                                    "MFD": np.array([])}}
        finally:
            result[rec.subject][rec.known]["MSA"] = np.append(result[rec.subject][rec.known]["MSA"], rec.MSA)
            result[rec.subject][rec.known]["MFD"] = np.append(result[rec.subject][rec.known]["MFD"], rec.MFD)
    return result


def process_result(res):
    '''
    From resulted dictionary computes needed averages and standard deviations.
    '''
    subject_results = []
    MFD_overall_true = np.array([])
    MFD_overall_false = np.array([])
    MSA_overall_true = np.array([])
    MSA_overall_false = np.array([])
    for subject in sorted(res):
        subject_results.append(SubjectResult(
            subject,
            np.average(res[subject]["true"]["MFD"]),
            np.std(res[subject]["true"]["MFD"]),
            np.average(res[subject]["false"]["MFD"]),
            np.std(res[subject]["false"]["MFD"]),
            np.average(res[subject]["true"]["MSA"]),
            np.std(res[subject]["true"]["MSA"]),
            np.average(res[subject]["false"]["MSA"]),
            np.std(res[subject]["false"]["MSA"])))
        MFD_overall_true = np.append(MFD_overall_true, res[subject]['true']['MFD'])
        MFD_overall_false = np.append(MFD_overall_false, res[subject]['false']['MFD'])
        MSA_overall_true = np.append(MSA_overall_true, res[subject]['true']['MSA'])
        MSA_overall_false = np.append(MSA_overall_false, res[subject]['false']['MSA'])
    overall_result = OverallResult(
        np.average(MFD_overall_true), np.std(MFD_overall_true),
        np.average(MFD_overall_false), np.std(MFD_overall_false),
        np.average(MSA_overall_true), np.std(MSA_overall_true),
        np.average(MSA_overall_false), np.std(MSA_overall_false))
    return subject_results, overall_result


def generate_result_csv(subject_results, overall_result):
    '''
    Creates result csv with all needed averages and standard deviations
    '''
    result = ""
    for subject_data in subject_results:
        line = " ".join(map(str, subject_data))
        line += " " + " ".join(map(str, overall_result))
        result += line + '\n'
    return result
```

## Results
Now we can move to actual per subject and overall results.
First we compute result dict that contains MSA and MFD per every subject and known state.

```python
rd = get_result_dict()
```

Now we compute averages and standard deviations from that dict.

```python
subjects_results, overall_results = process_result(rd)
```

We write those data to table.

```python
from IPython.core.display import HTML
html = ['<table>']
fields = subjects_results[0]._fields
html.append('<tr>')
html.extend(['<td><strong>{0}</strong></td>'.format(field) for field in fields])
html.append('</tr>')

for subject_result in subjects_results:
    html.append('<tr>')
    html.append('<td><strong>{0}</strong></td>'.format(subject_result.subject))
    html.extend(['<td>{:0.3f}</td>'.format(value) for value in subject_result[1:]])
    html.append('</tr>')
    
html.append('<tr>')
html.append('<td><strong>Overall</strong></td>')
html.extend(['<td>{:0.3f}</td>'.format(field) for field in overall_results])
html.append('</tr>')

html.append("</table>")
HTML('\n'.join(html))
```

And with those results we can create our bar charts with averages and standard deviations.

```python

for index, subject_result in enumerate(subjects_results):
    sd = False
    ys = np.array([])
    xs = np.arange(4)
    stds = np.array([])
    labels = []
    width = 0.8
    for name in subject_result._fields:
        if name == 'subject':
            continue
        if not sd:
            y = subject_result.__getattribute__(name)
            labels.append(name)
            sd = True
        else:
            std = subject_result.__getattribute__(name)
            sd = False
            ys = np.append(ys, y)
            stds = np.append(stds, std)
    
    plt.figure(0, figsize(14,14))
    ax1 = plt.subplot2grid((3, 4), (index // 2, 2 * (index % 2)))
    ax1.bar(xs[0:2], ys[0:2],   width, yerr=stds[0:2], color='b', ecolor='k')
    ax1.set_title(subject_result.subject + ' MFD')
    ax1.set_xticks(xs[0:2] + width / 2)
    ax1.set_xticklabels(labels[0:2])
    ax1.set_ylabel('Time [s]')
    
    ax2 = plt.subplot2grid((3, 4), (index // 2, 2 * (index % 2) + 1))
    ax2.bar(xs[2:], ys[2:], width, yerr=stds[2:], color='r', ecolor='k')  
    ax2.set_title(subject_result.subject + ' MSA')
    ax2.set_xticks(xs[2:] + width / 2)
    ax2.set_xticklabels( labels[2:])
    ax2.set_ylabel('Velocity')
    plt.tight_layout()
```

```python
sd = False
ys = np.array([])
xs = np.arange(4)
stds = np.array([])
labels = []
width = 0.8
for name in overall_results._fields:
    if not sd:
        y = overall_results.__getattribute__(name)
        labels.append(name)
        sd = True
    else:
        std = overall_results.__getattribute__(name)
        sd = False
        ys = np.append(ys, y)
        stds = np.append(stds, std)

fig = plt.figure(figsize(7,5))
ax = fig.add_subplot(121)
ax.bar(xs[0:2], ys[0:2],   width, color='b', yerr=stds[0:2], ecolor='k')
    
ax.set_title('Overall resutls MFD')
ax.set_xticks(xs[0:2] + width / 2)
ax.set_xticklabels(labels[0:2])
ax.set_ylabel('Time [s]')
    
ax2 = fig.add_subplot(122)
ax2.bar(xs[2:], ys[2:], width, color='r', yerr=stds[2:], ecolor='k')
    
ax2.set_title('Overall results MSA')
ax2.set_xticks(xs[2:] + width / 2)
ax2.set_xticklabels(labels[2:])
ax2.set_ylabel('Velocity')
fig.show()
plt.tight_layout()
```
